```python
x = df.drop('Bankrupt?',axis=1)
y = df['Bankrupt?']
```

```python
x.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

# Balancing the data
x_overSampled,y_overSampled = smote.fit_resample(xtrain,ytrain)
```

```python
len(xtrain), len(x_overSampled)
```

```python
from sklearn.preprocessing import MinMaxScaler
X_test_over = xtest.copy()

### Oversampled data ###
LABELS = X_overSampled.columns

## Scaling ##
mm_scaler = MinMaxScaler()

# Apply Min-Max Scaling
X_overSampled[LABELS] = mm_scaler.fit_transform(x_overSampled[LABELS])
X_test_over[LABELS] = mm_scaler.transform(X_test_over[LABELS])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import confusion_matrix
```

```python
list=[LinearRegression(),
      DecisionTreeRegressor(),RandomForestRegressor(), GradientBoostingRegressor(),
      SVR(),KNeighborsRegressor(n_neighbors=3)]
```

```python
def predict(model, model_name, X_overSampled, y_overSampled, X_test_over, ytest):
    model1=model.fit(X_overSampled,y_overSampled)
    ypred=model1.predict(X_test_over)
    ypred = [1 if p >= 0.5 else 0 for p in ypred]
    trainac=model1.score(X_overSampled,y_overSampled)
    testac=model1.score(X_test_over,ytest)
    print(model_name)
    print("Training Accuracy", trainac)
    print("Testing Accuracy",testac)

    results={'model':model_name,
            'Training Accuracy':trainac,
            'Testing Accuracy':testac,
             'Confusion_matrix':confusion_matrix(ytest, ypred)}
    print('Confusion Matrix:')
    print(confusion_matrix(ytest,ypred))

    return results
```

```python
all_results=[]
for model in list:
    model_name = model.__class__.__name__
    results=predict(model, model_name, xtrain, ytrain, xtest, ytest)
    all_results.append(results)
```

```python
df1=pd.DataFrame(all_results)
print(df1)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(df1['model'], df1['Training Accuracy'], label='Training Accuracy')
plt.bar(df1['model'], df1['Testing Accuracy'], label='Testing Accuracy')
plt.xlabel('model')
plt.ylabel('Accuracy')
plt.title('Training and Testing Accuracy of Models')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
